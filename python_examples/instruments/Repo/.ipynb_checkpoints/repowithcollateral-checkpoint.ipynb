```python
import json
import pandas as pd
import numpy as np
from datetime import datetime
import prettytable as pt
import os
import warnings
warnings.filterwarnings("ignore")



###################################################################################################
                                                                                                  #
from rhoova.Client import *                                                                       #   
#Register and get api key from https://app.rhoova.com/ for ClientConfig("api key", "api secret")  #
config = ClientConfig("", "")                                                                     #
api = Api(config)                                                                                 # 
                                                                                                  #
###################################################################################################


directory = os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir )
yielddatadirectory=directory+"/data/yielddata/bonddefiniton.csv"
marketdatadirectory=directory+"/data/marketdata/marketdata.csv"


bonddefiniton = pd.read_csv(yielddatadirectory)
marketdata = pd.read_csv(marketdatadirectory)

yieldddata=bonddefiniton.merge(marketdata[['valuationDate','isinCode','value']], left_on='isinCode', right_on='isinCode')
```

```python

positiondata ={
  "valuationDate": "2021-06-18",
  "settlementMoney": None,
  "termMoney": None,
  "repoSettlementDate": "2021-06-21",
  "repoDeliveryDate": "2021-06-23",
  "repoType": "Repo",
  "currency": "TRY",
  "dayCounter": "Actual360",
  "notional": 1000000,
  "marginOrHaircut": "Haircut",
  "marginOrHaircutRate": 1,
  "dirtyPrice": None,
  "exdividend": True,
  "repoRate": 0.00035,
  "collateral": {
    "settlementDate": "2021-06-18",
    "fixedRateBondDefinition": {
      "coupon": 0.09,
      "frequency": "Semiannual",
      "dayCounter": "ActualActual",
      "issueDate": "2014-08-06",
      "redemption": 100,
      "calendar": "Turkey",
      "maturityDate": "2024-07-24",
      "businessDayConvention": "ModifiedFollowing",
      "maturityDateConvention": "ModifiedFollowing",
      "dateGeneration": "Backward",
      "endOfMonth": False
    },
    "discountCurve": {
      "settlementDays": 0,
      "intpMethod": "Linear",
      "currency": "TRY",
      "calendar": "Turkey",
      "dayCounter": "ActualActual",
      "period": "6M",
      "instruments": {
        "BONDS": {
          "settlementDays": 2
        }
      }
    }
  },
  "yieldData": yieldddata.to_dict('r')
}
try:
    reporesult = api.createTask(CalculationType.REPO, positiondata,True)
    result=json.loads(reporesult["result"])
except RhoovaError as e:
    e.printPretty()    
```
