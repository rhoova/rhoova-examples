```python
data = {
  "notional": 10000000,
  "valuationDate": "2021-06-08",
  "settlementDate": "2021-06-10",
  "buySell": "buy",
  "fraDefinition": {
    "strike": 0.00127734,
    "frequency": "Quarterly",
    "startDate": "2021-08-10",
    "maturityDate": "2021-11-10",
    "businessDayConvention": "ModifiedFollowing",
    "dateGeneration": "Backward",
    "calendar": "UnitedStates",
    "dayCounter": "Actual360",
    "endOfMonth": True,
    "fixingDate": [],
    "fixingRate": [],
    "currency": "USD",
    "longShort": "Short"
  },
  "discountCurve": {
    "settlementDays": 2,
    "intpMethod": "linear",
    "currency": "USD",
    "calendar": "UnitedStates",
    "dayCounter": "Thirty360",
    "period": "3M",
    "instruments": {
      "IRS": {
        "frequency": "Semiannual",
        "businessDayConvention": "ModifiedFollowing",
        "dayCounter": "Thirty360"
      },
      "FUTURES": {
        "businessDayConvention": "ModifiedFollowing",
        "dayCounter": "Actual360",
        "market": "IMM",
      },
      "DEPO": {
        "businessDayConvention": "ModifiedFollowing",
        "dayCounter": "Actual360"
      }
    }
  },
  "yieldData":  yielddata.to_dict('r')
    }
```

```python
try:
    data = api.createTask(CalculationType.FORWARD_RATE_AGREEMENT, data, True)
    if(data["result"]):
        result=json.loads(data["result"])
    else:
        print(data["result"])
except RhoovaError as e:
    e.printPretty()
```
