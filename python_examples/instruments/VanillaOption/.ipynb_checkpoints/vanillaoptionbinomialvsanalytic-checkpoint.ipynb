```python
import json
import prettytable as pt
import matplotlib.pyplot as plt
import os 
import warnings
warnings.filterwarnings("ignore")

from rhoova.Client import *


###################################################################################################
                                                                                                  #
from rhoova.Client import *                                                                       #   
#Register and get api key from https://app.rhoova.com/ for ClientConfig("api key", "api secret")  #
config = ClientConfig("", "")                                                                     #
api = Api(config)                                                                                 # 
                                                                                                  #
###################################################################################################
```

```python
data ={
  "valuationDate": "2021-03-17",
  "settlementDate": "2021-03-17",
  "maturityDate": "2021-06-17",
  "notional": 1000000,
  "optionDefinition": {
    "underlying": "USD",
    "currency": "TRY",
    "callPut": "Call",
    "exerciseType": "European",
    "longShort": "Short",
    "buySell": "Buy",
    "strike":7.81,
    "underlyingPrice":7.5109,
    "volatility": 0.19473,
    "interestRate": 0.16732,
    "riskFreeRate":  0.0028,
    "processType": "BSMerton",
    "method": "Analytic",
    "calendar": "Turkey",
    "businessDayConvention": "ModifiedFollowing",
    "dayCounter": "Actual360",
    "optionStartDate": "2021-03-17",
    "optionEndDate": "2021-06-17",
    "timeSteps": 800,
    "timeGrid": 801
  }
}

```

```python
def optionprices(optiondata,method="Analytic",exerciseType="European",steps=100):
    optiondata['optionDefinition']['method']=method
    optiondata['optionDefinition']['timeSteps']=steps
    optiondata['optionDefinition']['exerciseType']=exerciseType
    try:
        resultdata = api.createTask(CalculationType.VANILLA_OPTION, optiondata,True)
        result=json.loads(resultdata["result"])
    except RhoovaError as e:
        e.printPretty()
    return result.get('pv')
```

```python
steps = range(2, 200, 1)
europeanbinomialprices = [optionprices(data,"Binomial Cox-Ross-Rubinstein","European",step) for step in steps]
```

```python
europeanbsprice=optionprices(data)
```

```python
europeanbsprices = [europeanbsprice for x in range(2, 200, 1)]
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

x=steps
y1 =europeanbinomialprices
y2 = europeanbsprices

ax.plot(x,y1, 'g', label='Binomial Price', linewidth=2)
ax.plot(x,y2, 'r', label='Analytic', linewidth=2)
#
leg = ax.legend(prop={'size': 16});
plt.grid(linestyle='dashdot')
plt.show()
```
