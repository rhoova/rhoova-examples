```python
import json
import pandas as pd
import numpy as np
from datetime import datetime
import prettytable as pt
from matplotlib import pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")



###################################################################################################
                                                                                                  #
from rhoova.Client import *                                                                       #   
#Register and get api key from https://app.rhoova.com/ for ClientConfig("api key", "api secret")  #
config = ClientConfig("", "")                                                                     #
api = Api(config)                                                                                 # 
                                                                                                  #
###################################################################################################


directory = os.path.normpath(os.getcwd() + os.sep + os.pardir)
yielddatadirectory=directory+"/data/yielddata/bonddefiniton.csv"
marketdatadirectory=directory+"/data/marketdata/marketdata.csv"


bonddefiniton = pd.read_csv(yielddatadirectory)
marketdata = pd.read_csv(marketdatadirectory)

data=bonddefiniton.merge(marketdata[['valuationDate','isinCode','value']], left_on='isinCode', right_on='isinCode')
```

```python
instruments={}
bonds={}
bonds["method"]="NelsonSiegel"
bonds["numberOfIterations"]=10000
bonds["accuracy"]=0.000001
bonds["simplexLambda"]=10
bonds["settlementDays"]=2
instruments["BONDS"]=bonds
instruments
```

```python
yieldCurve={}
yieldCurve["settlementDays"]=2
yieldCurve["calendar"]="NullCalendar" 
yieldCurve["currency"]="USD" # Currency is using filter yield data. Only same currency datas are filtered for bond calculation
yieldCurve["dayCounter"]="Thirty360E" # Default value is Actual360
yieldCurve["period"]="6M" # Can not use if instruments selected BONDS
yieldCurve["intpMethod"]="Linear" # if method is not null, it will not used
yieldCurve["instruments"]=instruments
yieldCurve
```

```python
ycresult ={
  "valuationDate": "2021-03-05",
  "yieldCurve": yieldCurve,
  "yieldData": data.to_dict('r')
}
```

```python
try:
    result = api.createTask(CalculationType.YIELD_CURVE, ycresult,True)
    nsresult=json.loads(result["result"])
except RhoovaError as e:
    e.printPretty()
```

```python
nszerorates=nsresult.get("zeroRates")
nszeros={}
for i in nszerorates:
    nszeros[datetime.strptime(i.get("date"), "%Y-%m-%d")]=i.get("rate")
nszeros=dict(sorted(nszeros.items()))  
```

```python
yieldCurve["instruments"]["BONDS"]["method"]="Svensson"
try:
    result = api.createTask(CalculationType.YIELD_CURVE, ycresult,True)
    svresult=json.loads(result["result"])
except RhoovaError as e:
    e.printPretty()
```

```python
svzerorates=svresult.get("zeroRates")
svzeros={}
for i in svzerorates:
    svzeros[datetime.strptime(i.get("date"), "%Y-%m-%d")]=i.get("rate")
svzeros=dict(sorted(svzeros.items()))  
```

```python
cbzerorates=cbresult.get("zeroRates")
cbzeros={}
for i in cbzerorates:
    cbzeros[datetime.strptime(i.get("date"), "%Y-%m-%d")]=i.get("rate")
cbzeros=dict(sorted(cbzeros.items()))  
```

```python
exzerorates=exresult.get("zeroRates")
exzeros={}
for i in exzerorates:
    exzeros[datetime.strptime(i.get("date"), "%Y-%m-%d")]=i.get("rate")
exzeros=dict(sorted(exzeros.items()))  
```

```python
pwzerorates=pwresult.get("zeroRates")
pwzeros={}
for i in pwzerorates:
    pwzeros[datetime.strptime(i.get("date"), "%Y-%m-%d")]=i.get("rate")
pwzeros=dict(sorted(pwzeros.items()))  
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

x=list(nszeros.keys()) 
y1 =list(nszeros.values()) 
y2 =list(svzeros.values()) 
y3 =list(cbzeros.values()) 
y4 =list(exzeros.values()) 
y5 =list(spzeros.values()) 
#y6 =list(pwzeros.values()) 

ax.plot(x,y1, 'green', label='NelsonSiegel', linewidth=2)
ax.plot(x,y2, 'orange', label='Svensson', linewidth=2)
ax.plot(x,y3, 'red', label='Cubic Splines', linewidth=2)
ax.plot(x,y4, 'blue', label='Exponential Splines', linewidth=2)
ax.plot(x,y5, 'yellow', label='Simple Polynomial(3)', linewidth=2)
#ax.plot(x,y6, 'o', label='Piecewise Linear', linewidth=2)


leg = ax.legend(prop={'size': 16});
plt.grid(linestyle='dashdot')
plt.show()
```
